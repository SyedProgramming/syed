That means, if one investor has invested multiple times in one startup, count one for that company. There are many errors in startup names. Ignore correcting all, just handle the important ones - Ola, Flipkart, Oyo and Paytm.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
data_set=pd.read_csv('/Users/jasyed/Documents/VISA/Learnings/Materials/MachineLearning/Coding_Ninjas_ML/startup_funding.csv',encoding='utf-8')

data_set['CityLocation'].dropna(inplace=True) 

def separateCity(data):
    return str(data).split('/')[0].strip()

city_list=['New Delhi','Bangalore','Mumbai','Gurgaon','Noida']
```

```python
# Cleanup City names
data_set['CityLocation'] = data_set['CityLocation'].replace("Delhi","New Delhi")
data_set['CityLocation'] = data_set['CityLocation'].replace("bangalore","Bangalore")

data_set['CityLocation'] = data_set['CityLocation'].apply(separateCity)


city_list_data_df = data_set[data_set.CityLocation.isin(city_list)]

city_list_data_df['StartupName'] = city_list_data_df['StartupName'].dropna() 
city_list_data_df['StartupName'] = city_list_data_df['StartupName'].replace("Flipkart.com","Flipkart")
city_list_data_df['StartupName'] = city_list_data_df['StartupName'].replace("Ola Cabs","Ola")
city_list_data_df['StartupName'] = city_list_data_df['StartupName'].replace("Olacabs","Ola")
city_list_data_df['StartupName'] = city_list_data_df['StartupName'].replace("Oyo Rooms","Oyo")
city_list_data_df['StartupName'] = city_list_data_df['StartupName'].replace("OYO Rooms","Oyo")
city_list_data_df['StartupName'] = city_list_data_df['StartupName'].replace("OyoRooms","Oyo")
city_list_data_df['StartupName'] = city_list_data_df['StartupName'].replace("Oyorooms","Oyo")
city_list_data_df['StartupName'] = city_list_data_df['StartupName'].replace("Paytm Marketplace","Paytm") 
```

```python
def createDictionary(array):
    dictionary={}
    
    for i in array:
        if ',' not in i:
            if i in dictionary:
                dictionary[i] = dictionary.get(i)+1
            else:
                dictionary[i] = 1
        else:
            string = i.split(',')
            for j in string:
                if j.strip() in dictionary:
                    dictionary[j.strip()] = dictionary.get(j.strip())+1
                else:
                    dictionary[j.strip()]=1
    return dictionary
```

```python
city_list_data_df['InvestorsName'].dropna(inplace=True)
city_list_data_df.head()

city_list_data_df_1 = city_list_data_df[['InvestorsName','StartupName']]
```

```python
#city_list_data_df_1.explode('InvestorsName')
city_list_data_df_1.info()
```

```python
city_list_data_df_1 = city_list_data_df_1.assign(InvestorsName=city_list_data_df_1['InvestorsName'].str.split(',')).explode('InvestorsName')
city_list_data_df_1.count()
```

```python
city_list_data_df_2 = city_list_data_df_1.groupby(['InvestorsName', 'StartupName'])['InvestorsName'].count().sort_values(ascending=False)

city_list_data_df_2_index = city_list_data_df_2.index
city_list_data_df_2_values = city_list_data_df_2.values
```

```python
city_list_data_df_3 = list(city_list_data_df_2_index.map(lambda x : x[0].split(',')))
city_list_data_df_3_flat = [item for i in city_list_data_df_3 for item in i]
city_list_data_df_3_flat
```

```python
dictionary={}
    
for i in city_list_data_df_3_flat:
    if i in dictionary:
        dictionary[i] = dictionary.get(i)+1
    else:
        dictionary[i] = 1
```
