#### Due to your immense help, your friend startup successfully got seed funding and it is on the operational mode. Now your friend wants to expand his startup and he is looking for new investors for his startup. Now you again come as a saviour to help your friend and want to create a list of probable new new investors. Before moving forward you remember your investor friend advice that finding the investors by analysing the investment type. Since your friend startup is not in early phase it is in growth stage so the best-suited investment type is Private Equity. Find the top 5 investors who have invested in a different number of startups and their investment type is Private Equity. Correct spelling of investment types are - "Private Equity", "Seed Funding", "Debt Funding", and "Crowd Funding". Keep an eye for any spelling mistake. You can find this by printing unique values from this column.There are many errors in startup names. Ignore correcting all, just handle the important ones - Ola, Flipkart, Oyo and Paytm.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
data_set=pd.read_csv('/Users/jasyed/Documents/VISA/Learnings/Materials/MachineLearning/Coding_Ninjas_ML/startup_funding.csv',encoding='utf-8')

data_set['CityLocation'].dropna(inplace=True) 

def separateCity(data):
    return str(data).split('/')[0].strip()

city_list=['New Delhi','Bangalore','Mumbai','Gurgaon','Noida']
investment_type=['Private Equity']
```

```python
# Cleanup City names
data_set['CityLocation'] = data_set['CityLocation'].replace("Delhi","New Delhi")
data_set['CityLocation'] = data_set['CityLocation'].replace("bangalore","Bangalore")

data_set['CityLocation'] = data_set['CityLocation'].apply(separateCity)


city_list_data_df_temp = data_set[data_set.CityLocation.isin(city_list)]
city_list_data_df_temp['StartupName'] = city_list_data_df_temp['StartupName'].dropna() 
city_list_data_df_temp['StartupName'] = city_list_data_df_temp['StartupName'].replace("Flipkart.com","Flipkart")
city_list_data_df_temp['StartupName'] = city_list_data_df_temp['StartupName'].replace("Ola Cabs","Ola")
city_list_data_df_temp['StartupName'] = city_list_data_df_temp['StartupName'].replace("Olacabs","Ola")
city_list_data_df_temp['StartupName'] = city_list_data_df_temp['StartupName'].replace("Oyo Rooms","Oyo")
city_list_data_df_temp['StartupName'] = city_list_data_df_temp['StartupName'].replace("OYO Rooms","Oyo")
city_list_data_df_temp['StartupName'] = city_list_data_df_temp['StartupName'].replace("OyoRooms","Oyo")
city_list_data_df_temp['StartupName'] = city_list_data_df_temp['StartupName'].replace("Oyorooms","Oyo")
city_list_data_df_temp['StartupName'] = city_list_data_df_temp['StartupName'].replace("Paytm Marketplace","Paytm") 




city_list_data_df_temp['InvestmentType'] = city_list_data_df_temp['InvestmentType'].replace("PrivateEquity","Private Equity") 
city_list_data_df_temp['InvestmentType'] = city_list_data_df_temp['InvestmentType'].replace("SeedFunding","Seed Funding")
city_list_data_df_temp['InvestmentType'] = city_list_data_df_temp['InvestmentType'].replace("Crowd funding","Crowd Funding") 


city_list_data_df = city_list_data_df_temp[city_list_data_df_temp.InvestmentType.isin(investment_type)]

```

```python
city_list_data_df['InvestmentType'].unique()
```

```python
city_list_data_df['InvestorsName'].dropna(inplace=True)
city_list_data_df_1 = city_list_data_df[['InvestorsName','StartupName']]
```

```python
city_list_data_df_1.info()
```

```python
city_list_data_df_2 = city_list_data_df_1.groupby(['InvestorsName', 'StartupName'])['InvestorsName'].count().sort_values(ascending=False)

city_list_data_df_2_index = city_list_data_df_2.index
city_list_data_df_2_values = city_list_data_df_2.values
```

```python
city_list_data_df_3 = list(city_list_data_df_2_index.map(lambda x : x[0].split(',')))
city_list_data_df_3_flat = [item for i in city_list_data_df_3 for item in i]
city_list_data_df_3_flat
```

```python
dictionary={}
    
for i in city_list_data_df_3_flat:
    if i in dictionary:
        dictionary[i] = dictionary.get(i)+1
    else:
        dictionary[i] = 1
```

```python
dataf = pd.DataFrame(list(dictionary.values()), list(dictionary.keys()))
dataf=dataf.sort_values(by=[0], ascending=False)
dataf.head(8)
```
