```python
setup = '''
import numpy as np
import pandas as pd
df = pd.read_csv('Data/tips.csv')
def quality(total_bill, tip):
    if tip/total_bill > 0.25:
        return "Generous"
    else:
        return "Other"
'''
```

```python
stmt_one = '''
df['Tip_Quality_1'] = df[['total_bill', 'tip']].apply(lambda df : quality(df['total_bill'], df['tip']), axis = 1)
'''

stmt_two = '''
df['Tip_Quality_2'] = np.vectorize(quality)(df['total_bill'], df['tip'])
'''
```

```python
timeit.timeit(setup=setup, stmt=stmt_one, number=1000)
```

```python
timeit.timeit(setup=setup, stmt=stmt_two, number=1000)
```
