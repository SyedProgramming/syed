## Histogram Workshop

```python
import matplotlib.pyplot as plt
```

```python
with open('salary.txt', 'r') as infile:
    data = [float(i) for i in infile.readlines()]
```

```python
plt.scatter(x=data, y=[1 for i in range(len(data))], alpha=0.05)
plt.show()
```

```python
plt.hist(data, bins=100)
plt.show()
```

```python
plt.hist(data, bins=bins)
plt.show()
```

```python
plt.hist(data, bins=bins, label="Salaries")
plt.xlabel("GBP")
plt.ylabel("Number of Jobs")
plt.legend()
plt.title("Histogram of Salary Data")
plt.xlim([0,100000])
plt.savefig("salary_histogram.png")
plt.show()
```

```python
plt.hist(data, bins=list(range(0, 10000, 500)), label="Salaries")
plt.xlabel("GBP")
plt.ylabel("Number of Jobs")
plt.legend()
plt.title("Histogram of Salary Data")
plt.xlim([0,100000])
plt.savefig("salary_histogram.png")
plt.show()
```

```python
plt.hist(data, bins=bins, cumulative=True, histtype='step', density=True)
plt.xlabel("GBP")
plt.ylabel("Number of Jobs")
plt.title("Histogram of Salary Data")
#plt.xlim([0,max(data)])
plt.savefig("salary_histogram.png")
plt.show()
```

```python
data = [37, 33, 33, 32, 29, 28, 28, 23, 22, 22, 22, 21, 21, 21, 20, 20, 19, 19, 18, 18, 18, 18, 16, 15, 14, 14, 14, 12, 12, 9, 6]

stem_leaf={}
for d in data:
    stem = d//10
    leaf = d%10
    if stem in stem_leaf:
        stem_leaf[stem].append(leaf)
    else:
        stem_leaf[stem] = [leaf]


for stem in sorted(stem_leaf, reverse=True):
    print(stem, end="  ")
    
    for leaf in sorted(stem_leaf[stem]):
        print(leaf, end = "   ")
    print('')
    
```
