## Support Vector Regression - Movies Data Collection Prediction

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Movie_regression.csv", header=0)
data.head()
```

```python
data.info()
```

### Missing Value Imputation

```python
data.describe().transpose()
```

```python
data['Time_taken'].mean()
```

```python
data['Time_taken'].fillna(value=data['Time_taken'].mean(), inplace=True)
```

```python
data.info()
```

### Dummy Variable Creation

```python
data.head()
```

```python
data = pd.get_dummies(data=data, columns=['3D_available','Genre'], drop_first=True)
```

```python
data.head()
```

### Splitting data into X and y

```python
X = data.loc[:,data.columns != 'Collection']
type(X)
```

```python
X.head()
```

```python
y = data.loc[:,data.columns == 'Collection']
type(y)
```

```python
y.head()
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2,random_state=0)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
X_test.head()
```

```python
y_test.head()
```

```python
print(X_train.shape)
print(y_train.shape)
print('---------')
print(X_test.shape)
print(y_test.shape)
```

### Standardizing Data
https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
X_train_std
```

```python
X_test_std
```

### Training Regression Tree
https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVR.html

```python
from sklearn.svm import SVR
svr = SVR(kernel='linear', C=3000)
```

```python
svr.fit(X_train_std, y_train)
```

### Predict values using trained model

```python
y_train_pred = svr.predict(X_train_std)
y_test_pred = svr.predict(X_test_std)
```

```python
y_test_pred
```

### Model Performance

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mean_squared_error(y_test, y_test_pred)
```

```python
mean_absolute_error(y_test, y_test_pred)
```

```python
r2_score(y_test, y_test_pred) * 100
```

```python
#RMSE
np.sqrt(mean_squared_error(y_test, y_test_pred))
```
