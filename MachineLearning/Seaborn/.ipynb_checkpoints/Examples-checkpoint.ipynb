```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dm_office_sales.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(12,5), dpi=200)
sns.scatterplot(x='salary',y='sales',data=df, hue='level of education');
```

## Scatter Plot

```python
plt.figure(figsize=(12,5), dpi=200)
sns.scatterplot(x='salary',y='sales',data=df, hue='salary');
```

```python
plt.figure(figsize=(12,5), dpi=200)
sns.scatterplot(x='salary',
                y='sales',
                data=df, 
                hue='level of education', 
                palette='Dark2',
                size='salary',
                alpha=0.5, #transparency
                style='level of education');
plt.savefig("my_plot.jpg")
plt.show()
```

## Distribution Plot

```python
df.head()
```

```python
plt.figure(figsize=(3,2), dpi=200)
sns.rugplot(df['salary'])
```

```python
plt.figure(figsize=(3,2), dpi=200)
#sns.set_style('whitegrid')
sns.distplot(df['salary'], 
             bins=100,
             color='red');
```

```python
plt.figure(figsize=(3,2), dpi=200)
#sns.set_style('whitegrid')
sns.distplot(df['salary'], 
             bins=100,
             color='red',
             hist=True,
             kde=False,
             rug=True);
```

```python
sns.kdeplot(df['salary'])
```
