## Figure Object

```python
a = np.linspace(0, 10, 11)
b = a**4
```

```python
a
```

```python
b
```

```python
x = np.arange(0,10)
y = x * 2
```

```python
x
```

```python
y
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y)
plt.show()
```

```python
fig = plt.figure()

# LARGE AXES
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel("A")
axes1.set_ylabel("B")
axes1.set_title("Power of 4")

# SMALL AXES
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25])
axes2.plot(x,y)
axes2.set_xlim(1,2)
axes2.set_ylim(0,25)
axes2.set_xlabel("X")
axes2.set_ylabel("Y")
axes2.set_title("Zoomed In")


plt.show()
```

```python
type(fig)
```

```python
fig = plt.figure(figsize=(8,5),dpi=100)

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

fig.savefig("new_figure.png", bbox_inches='tight')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(4,6), dpi=100)

axes[0][0].plot(x,y)
axes[0][1].plot(x,y)

axes[1][0].plot(x,y)
axes[1][0].set_xlabel("X")
axes[1][0].set_ylabel("Y")
axes[1][0].set_title("Title")

axes[1][1].plot(a,b)
axes[1][1].set_xlim(0,6)

fig.suptitle("Figure Level Title", fontsize=12)

plt.tight_layout()
#fig.subplots_adjust(wspace=5.0, hspace=1.0)

fig.set_figwidth(10)
fig.savefig("new_subplots.png", bbox_inches="tight")
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0, 1,1])

ax.plot(x,y, label="xy", color='green', lw=2, ls='--', marker='s', ms=10,
       markerfacecolor='red', markeredgewidth=5, markeredgecolor='orange');
ax.plot(a,b, label="ab", color='#FF334F', linewidth=4, linestyle='-.'); #RGB Hex Code

#ax.legend(loc=4)
ax.legend(loc=(0.5,0.5))
```
